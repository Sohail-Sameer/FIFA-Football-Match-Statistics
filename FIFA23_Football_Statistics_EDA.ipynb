# **3. Data Wrangling**

```python
# Check for missing values
df.isnull().sum()
```

```python
# Check for duplicate values
df.duplicated().sum()
```

```python
# Check datatypes
df.dtypes
```

```python
# ── Clean the Position column (remove HTML span tags) ──────────────────────
df['Position_clean'] = df['Position'].str.extract(r'pos\d+">(\w+)')

# ── Convert Value column to numeric (€ symbols, M/K suffixes) ─────────────────
def parse_currency(val):
    if pd.isnull(val):
        return np.nan
    val = str(val).replace('€','').strip()
    if val.endswith('M'):
        return float(val[:-1]) * 1_000_000
    elif val.endswith('K'):
        return float(val[:-1]) * 1_000
    else:
        try:
            return float(val)
        except:
            return np.nan

df['Value_num'] = df['Value'].apply(parse_currency)
df['Wage_num']  = df['Wage'].apply(parse_currency)
df['Release_Clause_num'] = df['Release Clause'].apply(parse_currency)

# ── Convert Height to cm (format: e.g. "181cm") ────────────────────────────────
df['Height_cm'] = df['Height'].str.extract(r'(\d+)cm').astype(float)

# ── Convert Weight to kg (format: e.g. "77kg") ─────────────────────────────────
df['Weight_kg'] = df['Weight'].str.extract(r'(\d+)kg').astype(float)

print("Cleaned columns created: Position_clean, Value_num, Wage_num, Release_Clause_num, Height_cm, Weight_kg")
df[['Name','Overall','Value_num','Wage_num','Position_clean','Height_cm','Weight_kg']].head()
```

The data wrangling process involved cleaning and transforming the dataset into a format suitable for exploratory data analysis. The following steps were performed:

- **Missing Values:** The dataset has missing values primarily in `Club` (211), `Joined` (1098), `Release Clause` (1151), and `Loaned From` (16966 — most players are not on loan, so this is expected and not a concern). These columns are either unused or handled appropriately in visualisations.
- **Duplicate Values:** The dataset has no duplicate rows.
- **Column Cleaning:** The `Position` column contained HTML span tags which were extracted using regex. The `Value`, `Wage`, and `Release Clause` columns stored currency as strings and were converted to numeric float values (in euros). `Height` and `Weight` were also parsed to numeric formats (cm and kg respectively).

No rows were dropped. The dataset is now clean and ready for analysis.


# **4. Data Visualisation, Storytelling & Experimenting with Charts : Understand the relationship between variables**

## Chart - 1

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(df['Overall'], bins=30, color='#1a6faf', edgecolor='white', linewidth=0.6)
ax.set_title('Distribution of Player Overall Ratings', fontsize=15, fontweight='bold', pad=12)
ax.set_xlabel('Overall Rating', fontsize=12)
ax.set_ylabel('Number of Players', fontsize=12)
ax.axvline(df['Overall'].mean(), color='#e74c3c', linestyle='--', linewidth=1.8, label=f"Mean: {df['Overall'].mean():.1f}")
ax.legend(fontsize=11)
sns.despine()
plt.tight_layout()
plt.show()
```

### 1. Why did you pick this specific chart?

##### Ans: A histogram was chosen to visualise the frequency distribution of overall player ratings, which is the primary performance metric in the dataset. It effectively shows the shape, spread, and central tendency of ratings across all players.

### 2. What is/are the insight(s) found from the chart?

##### Ans: The histogram reveals that overall player ratings follow a roughly bell-shaped (near-normal) distribution, with the majority of players rated between 60 and 75. The mean rating is approximately 66. Very few players achieve elite ratings above 85, confirming that top-tier talent is rare. The distribution is slightly left-skewed, indicating a long tail of lower-rated players at the 43–55 range.

### 3. Will the gained insights help creating a better machine learning model?

##### Ans: Yes. Understanding the rating distribution helps in deciding whether to normalise or standardise the Overall column before modelling. The skew and range also inform whether regression or classification approaches are more appropriate for predicting player ratings or market value.

## Chart - 3

```python
# Consider only nationalities with at least 50 players for reliability
nation_avg = (df.groupby('Nationality')['Overall']
               .agg(['mean','count'])
               .query('count >= 50')
               .sort_values('mean', ascending=False)
               .head(10))

fig, ax = plt.subplots(figsize=(10, 5))
palette = sns.color_palette('RdYlGn', 10)
bars = ax.bar(nation_avg.index, nation_avg['mean'], color=palette, edgecolor='white', linewidth=0.6)
ax.set_title('Top 10 Nations by Average Overall Rating\n(min. 50 players)', fontsize=15, fontweight='bold', pad=12)
ax.set_xlabel('Nationality', fontsize=12)
ax.set_ylabel('Average Overall Rating', fontsize=12)
ax.set_ylim(60, 74)
for bar, val in zip(bars, nation_avg['mean']):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1,
            f'{val:.1f}', ha='center', fontsize=10, fontweight='bold')
plt.xticks(rotation=30, ha='right')
sns.despine()
plt.tight_layout()
plt.show()
```

### 1. Why did you pick this specific chart?

##### Ans: A bar chart was selected to compare average overall ratings across nations. Restricting to nations with at least 50 players ensures the averages are statistically reliable and not skewed by a single star player from a small nation.

### 2. What is/are the insight(s) found from the chart?

##### Ans: The chart reveals that despite England having the most players, it does not produce the highest average ratings. Nations like Spain, France, Portugal, and Belgium tend to have higher average player quality. This suggests that some nations prioritise quality over quantity in professional football, and that European football academies (especially in Spain and France) are producing elite-level talent consistently.

### 3. Will the gained insights help creating a better machine learning model?

##### Ans: Yes. Average nationality rating is a strong feature indicator. It can be used as an aggregated feature (mean encoding) in models predicting player value or overall rating, reducing cardinality while retaining predictive signal.

### 2. What is/are the insight(s) found from the chart?

##### Ans: The chart clearly shows that younger players (lighter/yellow points) tend to have a larger gap above the diagonal, meaning they have more room to grow. ### 1. Why did you pick this specific chart?

##### Ans: A heatmap was chosen to display the interaction between two ordinal categorical variables — Skill Moves and International Reputation — and how they jointly relate to Average Overall Rating. This matrix view is far more compact and insightful than plotting them separately.

## Chart - 10

```python
# Select relevant numeric columns for correlation analysis
num_cols = ['Age', 'Overall', 'Potential', 'Special', 'International Reputation',
            'Weak Foot', 'Skill Moves', 'Value_num', 'Wage_num', 'Height_cm', 'Weight_kg']

corr_df = df[num_cols].dropna()

fig, ax = plt.subplots(figsize=(11, 8))
mask = np.triu(np.ones_like(corr_df.corr(), dtype=bool))
sns.heatmap(corr_df.corr(), annot=True, fmt='.2f', cmap='coolwarm',
            mask=mask, ax=ax, linewidths=0.5,
            cbar_kws={'label': 'Pearson Correlation Coefficient'},
            annot_kws={'size': 9})
ax.set_title('Correlation Heatmap — FIFA 23 Player Attributes', fontsize=15, fontweight='bold', pad=12)
plt.xticks(rotation=45, ha='right', fontsize=9)
plt.yticks(fontsize=9)
plt.tight_layout()
plt.show()
```

### 1. Why did you pick this specific chart?

##### Ans: A lower-triangle correlation heatmap was chosen as the final chart because it provides a comprehensive view of all pairwise relationships between numerical variables simultaneously. This helps confirm insights found in earlier charts and reveals hidden correlations not explored individually. The mask removes the redundant upper triangle for cleaner presentation.

### 2. What is/are the insight(s) found from the chart?

##### Ans: The heatmap reveals several important correlations. **Overall and Wage_num** have the strongest positive correlation (≈0.79), confirming that performance directly drives earnings. **Overall and Value_num** also show strong positive correlation (≈0.72), validating Overall as a key market pricing signal. **Special and Overall** are highly correlated (≈0.93), which makes sense as Special is a sum of all individual attributes. **Age and Value_num** show a negative correlation, meaning younger players command higher market values. **Skill Moves and Overall** have a moderate positive correlation, while **Height_cm and Weight_kg** are positively correlated as expected. **Age and Potential** are negatively correlated, confirming the earlier finding that older players are closer to their ceiling.

### 3. Will the gained insights help creating a better machine learning model?

##### Ans: Yes. The correlation matrix is an essential preprocessing tool for machine learning. High correlations between Special and Overall suggest multicollinearity, meaning both should not be used as independent features simultaneously. The strong predictors of Value_num (Overall, Wage, Age) identified here can be used as the primary feature set for a regression model predicting player market value.

The FIFA 23 Football Player Statistics EDA project successfully uncovered key patterns and relationships within professional football player data. Through 10 focused visualisations and a detailed correlation analysis, several important findings emerged:

1. Player Overall ratings follow a near-normal distribution with a mean of approximately 66, with elite talent (85+) being rare.
2. England produces the most players but not the highest-rated ones — Spain, France and Belgium lead in average quality.
3. Most professional players are aged 21–25, and age negatively correlates with growth potential.
4. Younger players show a larger gap between Potential and Overall ratings, identifying them as high-growth investments.
5. Left-footed players are significantly rarer (23%) yet show marginally higher average ratings.
6. Weekly Wage grows exponentially with Overall rating, with elite players earning disproportionately high salaries.
7. Both Skill Moves and International Reputation are strong predictors of Overall rating, with Reputation having the stronger effect.
8. Central positions (GK, CM, CB) tend to attract the highest-rated players in professional football.
9. The correlation heatmap confirms Overall, Wage, and Age as the three strongest predictors of player market Value.

Overall, this project demonstrates how exploratory data analysis can transform raw football statistics into actionable insights for scouting, squad building, and market strategy in professional football.
